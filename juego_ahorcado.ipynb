CUERPO PRINCIPAL DEL JUEGO:

# Ahorcado

- Reglas del juego del "Ahorcado":

    - Un jugador elige una palabra secreta y dibuja un espacio para cada letra de la palabra.

    - El otro jugador intenta adivinar letras para completar la palabra.

    - Si el jugador adivinador adivina una letra correctamente, se revela en su lugar correspondiente.

    - Si el jugador adivinador adivina incorrectamente, se dibuja una parte del cuerpo en la horca.

    - El objetivo del jugador adivinador es adivinar la palabra antes de que se dibuje el dibujo completo en la horca.


- Pistas para las alumnas:

    1. Empieza por seleccionar una palabra secreta para que el jugador adivinador intente adivinar. Puedes crear una lista de palabras.

    2. Inicializa el tablero de juego con guiones bajos ("_") para representar las letras no adivinadas.

    3. Utiliza un bucle `while` para permitir que el jugador adivinador haga múltiples intentos hasta que adivine la palabra o se agoten los intentos.

    4. Verifica si la letra adivinada por el jugador está en la palabra secreta y actualicen el tablero en consecuencia.

    5. Lleva un seguimiento de las letras adivinadas y los intentos fallidos.

    6. Determina las condiciones de victoria y derrota del juego, y salgan del bucle cuando se cumplan estas condiciones.


```python
# Elegimos 'import random' para elegir una palabra de la lista de forma aleatoria. 
import random 

# Lista de palabras en la que la máquina elige de manera aleatoria una de ellas:

palabras =["perro","fruta","murcielago","ornitorrinco"] # lista de palabras.
palabra_secreta = random.choice(palabras) # palabra secreta elegida por la máquina de manera aleatoria.
palabra_raya= '_' * len(palabra_secreta) # la palabra secreta elegida se imprimirá con guiones para que sea más visual.
intentos = 0 # crear una variable para contabilizar el número de intentos.

# Aquí empieza el juego. Mensaje de bienvenida e instrucciones:

print(" O    ___________________________________    O")
print("/|\\   ¡BIENVENID@  AL JUEGO DEL AHORCADO!   /|\\")
print("/ \\   ___________________________________   / \\")
print(" ")
print("¿COMO SE JUEGA?")
print("Tu como jugador/a debes adivinar la palabra letra por letra hasta formar dicha palabra en su totalidad.\n¡CUIDADO! Solo tendras un maximo de 6 INTENTOS para poder adivinarla, ya que cada vez que falles, se irá añadiendo una parte del cuerpo al ahorcado.\nSi el dibujo del ahorcado se completa antes de que hayas adividado la papabra secreta ---> PIERDES.\nSi adivinas la palabra secreta al completo antes de que se dibuje el ahorcado en su totalidad ---> GANAS.")
print("                                                   ")
print("¿List@? Empezamos en...... 3 ...... 2 ...... 1 ....... ¡A POR TODAS!")
print(" ")


while intentos < 6: # mientras el número de intentos sea menor que seis, la condición será verdadera y el bucle seguirá corriendo.
     
    print(f" La palabra secreta es:  {palabra_raya}  ")  
    adivina_letra = input("Elige una letra")
    
    if adivina_letra in palabra_secreta:

        print(f"Bien, has encontrado la letra '{adivina_letra}") 
        letra_adivinada_palabra = "" # nueva variable para ir formando la palabra secretra con las letras elegidas
        
        for letra in palabra_secreta: # itera en cada letra de la palabra secreta.
             if letra == adivina_letra or letra in palabra_raya: # si la letra de la palabra coincide con la letra elegida por el jugador.
                 letra_adivinada_palabra += letra                # si es así, esa letra se añadirá a la variable letra_adivinada_palabra.
                 
             else:
                letra_adivinada_palabra += "_"                   #si no es así, se añadirá un guion.
            
        palabra_raya = letra_adivinada_palabra # se actualiza el valor de la palabra_raya con las letras adivinadas que se han ido añadiendo a la variable letra_adivinada_palabra.


        if palabra_raya == palabra_secreta:
            print(f"¡ENHORABUENA! Efectivamente, la palabra secreta es: {palabra_secreta}. ¡BIEN JUGADO!")
            break # porque la condicion "palabra_raya == palabra_secreta" se ha cumplido, por lo que el juego termina y detiene el bucle.
    
       
        
    else: #si la letra no está en la palabra secreta, se van imprimiendo las partes del cuerpo del ahorcado:
        
        
        intentos += 1 # cada vez que el jugador introduzca una letra errónea, se incrementará el número de intentos en la variable hasta que llegue a 6 (su máximo) y la condición "intentos > 6" se volverá falsa.
        print(f"Lo siento, letra '{adivina_letra}' no encontrada. Estado de tu ahorcado: (Llevas {intentos} fallo(s), recuerda tienes hasta 6 fallos):")

        if intentos == 1:
            print("O")
        elif intentos == 2:
            print("O")
            print("|")
        elif intentos == 3:
            print(" O")
            print("/|")
        elif intentos == 4:
            print(" O")
            print("/|\\")
        elif intentos == 5:
            print(" O")
            print("/|\\")
            print("/")
        elif intentos == 6:
            print(" O")
            print("/|\\")
            print("/ \\")
            print("_____________________________________________")
            print(f"¡PERDISTE! La palabra secreta es ----> {palabra_secreta}")
            
            break #para salir del programa/ bucle.

```
